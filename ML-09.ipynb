```python
# Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may
# impact your analysis.
# Answer :-
import pandas as pd

# Load the Google Play Store dataset into a Pandas DataFrame
# Replace 'your_dataset.csv' with the actual file path or URL of the dataset
df = pd.read_csv('your_dataset.csv')

# Check for missing values
missing_values = df.isnull().sum()

# Print the columns with missing values and the count of missing values for each column
print("Columns with Missing Values:")
for column, count in missing_values.items():
    if count > 0:
        print(f"{column}: {count} missing values")

# The impact of missing values on your analysis can include:

# Data Completeness: Missing values can lead to incomplete datasets, potentially reducing the amount of data available for analysis.

# Biased Analysis: If missing values are not handled properly, your analysis may be biased, as it may only reflect the available data.

# Modeling Issues: Missing values can pose challenges for machine learning models that require complete data, potentially leading to model errors or suboptimal performance.

# Inaccurate Insights: Missing data can result in inaccurate insights and conclusions, particularly if the missing data is not missing completely at random (MCAR).

# To address missing values, you have several options, including:

# Imputation: Fill in missing values with reasonable estimates (e.g., mean, median, mode) based on the nature of the data.
# Deletion: Remove rows or columns with missing values, but exercise caution to avoid excessive data loss.
# Advanced imputation techniques: Use more advanced imputation methods, such as regression imputation or K-nearest neighbors imputation.
# Creating a missing data indicator: Create a binary indicator variable to flag missing data, which can be used as a feature in modeling.
# Your approach to handling missing values should depend on the specific dataset, the nature of the missing values, and the goals of your analysis. It's essential to document any missing data handling procedures to ensure transparency in your analysis.
```

```python
# Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
# the relationship.
# Answer :-
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the Google Play Store data

# Extract the columns containing app ratings and app sizes
rating_column = 'Rating'  # Replace with the actual column name for app ratings
size_column = 'Size'  # Replace with the actual column name for app sizes

# Create a scatter plot to visualize the relationship between app size and rating
plt.figure(figsize=(10, 6))
plt.scatter(df[size_column], df[rating_column], alpha=0.5, color='blue')
plt.title('App Size vs. Rating')
plt.xlabel('App Size')
plt.ylabel('Rating')
plt.grid()

plt.show()
```

```python
# Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.
# Answer :-
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the Google Play Store data

# Extract the columns containing app types and app prices
type_column = 'Type'  # Replace with the actual column name for app types (Free or Paid)
price_column = 'Price'  # Replace with the actual column name for app prices

# Calculate the average prices by app type
average_prices_by_type = df.groupby(type_column)[price_column].mean().reset_index()

# Create a bar chart to compare average prices by app type
plt.figure(figsize=(8, 6))
plt.bar(average_prices_by_type[type_column], average_prices_by_type[price_column], color='skyblue')
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
# with the highest number of installs.
# Answer :-
import pandas as pd

# Assuming 'df' is your DataFrame containing the Google Play Store data

# Extract the column containing app names and the number of installs
app_name_column = 'App'  # Replace with the actual column name for app names
installs_column = 'Installs'  # Replace with the actual column name for the number of installs

# Create a frequency table for app installs
install_counts = df.groupby(app_name_column)[installs_column].sum()

# Sort the frequency table in descending order to get the most popular apps
top_10_apps = install_counts.sort_values(ascending=False).head(10)

# Print the top 10 most popular apps
print(top_10_apps)
```
